```python
from faker import Faker
import pandas as pd
import random
```

```python
faker = Faker()
```

```python
def faker_data(data_you_want):
    try:
        return eval(f"[faker.{data_you_want}().split()[1] for i in range(10)]")
    except:
        try:
            return eval(f"[faker.{data_you_want}() for i in range(10)]")
        except:
            pass
```

```python
product_code = [random.randint(1000, 9999) for i in range(10)]
stock_qty = [random.randint(3, 200) for i in range(10)]
cost = [random.randint(10, 1000) for i in range(10)]
price = [random.randint(30, 2000) for i in range(10)]
last_week_sales = [random.randint(10, 50) for i in range(10)]
last_month_sales = [random.randint(10, 50) for i in range(10)]
```

```python
vendedores = faker_data("name")[:4]
vededores_series = [random.choice(vendedores) for i in range(10)]
```

```python
group = ["PG1", "PG2", "PG3", "PG4", "PG5"]
```

```python
product_group = [random.choice(group) for i in range(10)]
```

```python
df = pd.DataFrame({"vendedores": vededores_series, 
                   "product_group": product_group,
                  "product_code": product_code,
                  "stock_qty": stock_qty,
                  "cost": cost,
                  "price": price,
                  "last_week_sales": last_week_sales,
                  "last_month_sales": last_month_sales})
```

```python
df
```

#### Example 1: Single agregation

```python
df.groupby("product_group")["stock_qty"].mean()
```

#### Example 2: Multiple agregation

```python
df.groupby("product_group")["stock_qty", "price"].mean()
```

#### Example 3: Multiple agregation II

```python
df.groupby("product_group")["stock_qty"].agg(["min", "mean", "max", "std"])
```

#### Example 4a: Named agregations

```python
df.head(2)
```

```python
df_group_ex4 = df.groupby("product_group").agg(
                                stock_qtd = ("stock_qty", "median"),
                                avg_stock_qtd = ("stock_qty", "mean"),
                                max_stock_qtd = ("stock_qty", "max"))
df_group_ex4
```

#### Example 4b: Named agregations with insert

```python
df_group_ex4.insert(loc = len(df_group_ex4.columns), column = "mínimo", value = df.groupby("product_group").agg(stock_qtd_min = ("stock_qty", "min")))
df_group_ex4
```
