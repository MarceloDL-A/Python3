### Merge

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'lkey': ['foo', 'bar3', 'baz7', 'foo', 'baz'],
                    'value': [1, 2, 3, 5, 3]})
df2 = pd.DataFrame({'rkey': ['foo2sasa', 'barsasa1', 'baz', 'foo', 1],
                    'value': [5, 6, 7, 8, 1]})
```

```python
display(df1), display(df2)
```

```python
df1.merge(df2, left_on='lkey', right_on='rkey',
          suffixes=('_left', '_right'))
```

```python
left = pd.DataFrame(
    {
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
        "key": ["K0", "K1", "K0", "K1"],
    }
)


right = pd.DataFrame({"C": ["C0", "C1"], "D": ["D0", "D1"]}, index=["K0", "K1"])
```

```python
display(left)
display(right)
```

```python
result = pd.merge(
    left, right, left_on="key", right_index=True, how="left", sort=False
)
```

```python
result
```
